```python
from pathlib import Path
```

```python
from langchain_community.document_loaders import PyMuPDFLoader

Path = r'C:/Users/otavi/OneDrive/Desktop/mind_over_muscle.pdf'

loader = PyMuPDFLoader(Path)
docs = loader.load()

print(len(docs))
print(docs[0].page_content[:500])
print(docs[0].metadata)
```

```python
pip install langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=250,
)

chunks = splitter.split_documents(docs)
```

```python
for i, chunk in enumerate(chunks[10:15]):
    print(f"\n--- CHUNK {i} ---")
    print(chunk.page_content[:1500])
```

```python
pip install langchain-chroma langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="chroma_db"
)
```

```python
query = "Qual é o tema principal do livro?"

results = db.similarity_search(query, k=3)

for i, doc in enumerate(results):
    print(f"\n--- RESULTADO {i + 1} ---")
    print(doc.page_content[:1000])
    print(doc.metadata)
```

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="qwen2.5:0.5b",
    temperature=0
)
```

```python
query = "Quem foi Jigoro Kano?"

results = db.similarity_search(query, k=3)

contexto = "\n\n".join(
    f"[Trecho {i+1}]\n{doc.page_content}"
    for i, doc in enumerate(results)
)

prompt = f"""
Você é um assistente que responde perguntas sobre um livro.

Use apenas os trechos abaixo para responder.
Se a resposta não estiver nos trechos, diga que não encontrou essa informação nos trechos recuperados.

Trechos:
{contexto}

Pergunta:
{query}

Resposta:
"""

resposta = llm.invoke(prompt)

print(resposta.content)
```
